```python
import pandas as pd
import numpy as np
import datetime as dt
import warnings
warnings. filterwarnings('ignore')

import matplotlib.pyplot as plt
```

```python
# Scrapping NIFTY50 Stock tickers
stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[2].Symbol
stocks = stocks + '.NS'
stocks = stocks.to_list()
stocks
```

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
# NIFTY50 Stock closing price of last 5 years
df = yf.download(stocks, start='2019-01-01')['Close']
```

```python
ret_df = df.pct_change()
```

```python
# Yearly return
yearly_ret = (ret_df + 1).resample('Y').prod()
yearly_ret.T
```

```python
# calculate CAGR
def cagr(start_value, end_value, num_periods):
    return (end_value / start_value) ** (1 / (num_periods - 1)) - 1

start=df.loc['2019-1-1']
end=df.loc['2024-1-1']
year=5
CAGR=cagr(start,end,year)
CAGR.nlargest(10)
```

```python
# visualize the top 10 financial institution
CAGR.nlargest(10).plot.bar(figsize=(10,6))
plt.xticks(rotation=45)
plt.ylabel('CAGR (5 Years)')
plt.show()
```
